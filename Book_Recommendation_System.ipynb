# Data Preparation

```python
ds2=ds.copy()
#copying the dataframe
```

The use of loc attribute in order to access a collection of rows and columns using a boolean array of average rating and rating 

```python
ds2.loc[(ds2['average_rating']>=0)&(ds2['average_rating']<=1),'rating between']="between 0 and 1"
ds2.loc[(ds2['average_rating']>1)&(ds2['average_rating']<=2),'rating between']="between 1 and 2"
ds2.loc[(ds2['average_rating']>2)&(ds2['average_rating']<=3),'rating between']="between 2 and 3"
ds2.loc[(ds2['average_rating']>3)&(ds2['average_rating']<=4),'rating between']="between 3 and 4"
ds2.loc[(ds2['average_rating']>4)&(ds2['average_rating']<=5),'rating between']="between 4 and 5"
```

the function in pandas library which is sorting the dataframe in descending order of the provided values 

```python
ds2.sort_values('ratings_count')
```

here the dummies variable specifies whether or not both of the categorical variable takes on a particular values. it is also called dummies encoding 

it changes the variable into binary form that is 0 and 1

```python
rating_ds=pd.get_dummies(ds2['rating between'])

#generating the dummies data
```

the concatenation combines and joins together the two variables which in this condition is average rating and rating together 

```python
features=pd.concat([rating_ds, ds2['average_rating'], ds2['ratings_count']], axis=1)
features
```

# Building Book Recommendation System

importing from the sklearn library using min max scaler

```python
from sklearn.preprocessing import MinMaxScaler
```

the min max scaler is used to scale each feature to a specific range in order to transform it. here the estimator scales and translates each feature independently which falls inside the training set's given that the range such as 0 to 1 

```python
minmax=MinMaxScaler()

features=minmax.fit_transform(features)

features
```

importing of neighbor library and the model selection which is used to split test and train data 

```python
from sklearn import neighbors
from sklearn.model_selection import train_test_split
```

the use of model of kneighbours in the dataset which along with the ball tree algorithm 

the model fit refers to how well a machine learns and models generalizes the data which is comparable to the data it was trained on. 

```python
model=neighbors.NearestNeighbors(n_neighbors=9, algorithm='ball_tree')

model.fit(features)

dist, idlist=model.kneighbors(features)
```

```python
dist 
#showing of the model
```

```python
idlist #the listing of the models in the array 
```

final process to recommend the books in the list along with its title which is shown with the initial index of the book where the append fucntion adds the single ietm to an existing list. 


```python
def app(name):
    book_list=[]
    book_id=ds2[ds2['name']==name].index
    book_id=book_id[0]
    for new in idlist[book_id]:
        book_list.append(ds2.loc[new].name)
    return book_list
```

# Result

The final outcomes of the book recommendation is shown below.

```python
def app(name):
    book_list = []
    book_id = ds2[ds2['name'] == name].index
    if not book_id.empty:
        book_id = book_id[0]
        for new in idlist[book_id]:
            book_list.append((ds2.loc[new].name, ds2.loc[new].category))
    else:
        print("No book found with the given name.")
    return book_list

Books_names = app('Play Therapy')

# Print book names and categories
for name, category in Books_names:
    print(f"Book: {name}, Category: {category}")
```
